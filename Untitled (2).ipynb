```python
#Q. 1) For 'House of Cards', what is the Show Id and Who is the Director of this show ?
hfc = df[df['title'] == 'House of Cards']
s_id= hfc['show_id'].values[0]
dire = hfc['director'].values[0]

print("Show ID:", s_id)
print("Director:", dire)
```

```python
#Q. 2) In which year the highest number of TV Shows & Movies were released? Show with
Bar Graph.
yc= df.groupby('release_year')['title'].count()
yc
```

```python
plt.bar(yc.index, yc.values)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Number of TV Shows and Movies Released by Year')
plt.show()
```

```python
#Q. 4) Show all the Movies that were released in the year 2000.
m_count = df[df['type'] == 'Movie'].shape[1]
ts_count = df[df['type'] == 'TV Show'].shape[0]
print("Number of Movies:", m_count)
print("Number of TV Shows:", ts_count)
```

```python
m_2000 = df[df['release_year'] == 2000]['title']
print("Movies released in 2000:")
for movie in m_2000:
    print(movie)
```

```python
#Q. 5) Show only the Titles of all TV Shows that were released in India only.
tv_india = df[df['country'] == 'India']['title']
print("tv shows released in india:")
for tv in tv_india:
    print(tv)
```
